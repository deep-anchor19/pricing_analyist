```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("cuisine_customer.csv")
```

```python
df
```

```python
df_food = df.groupby(by=["cuisine","Customer_id"]).sum()
```

```python
df_food.reset_index(inplace=True)
```

```python
df_food
```

```python
df_food = df_food.pivot_table('Numbers of order',['cuisine'],'Customer_id')
```

```python
df_food.head(10)
```

```python
df_food.to_csv('processed_data.csv)
```
