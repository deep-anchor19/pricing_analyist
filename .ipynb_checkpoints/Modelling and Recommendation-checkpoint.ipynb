## 3.3 Recommendation - Whisky Flavour Profile<a class="anchor" id="3_3"></a>

### The different clusters of flavour profiles:

#### Single Malt and Blended Whiskies:

0. Unspecified - mostly blended whiskies
1. Full-bodied, sweet, pronounced sherry – with fruity, honey and spicy notes
2. Full-bodied, sweet, pronounced sherry – with fruity, floral and malty notes, some honey and spicy notes
3. Full-bodied, sweet, pronounced sherry – with fruity, floral, nutty, and spicy notes, some smoky notes
4. Medium-bodied, medium-sweet – with fruity, honey, malty and winey notes, some smoky and spicy notes
5. Full-bodied, sweet and malty – with fruity, spicy, and smoky notes
6. Light-bodied, sweet, apéritif-style – with honey, floral, fruity and spicy notes, but rarely any smoky notes
7. Very light-bodied, sweet, apéritif-style – with malty, fruity and floral notes 
8. Medium-bodied, medium-sweet, quite smoky – with some medicinal notes and spicy, fruity and nutty notes
9. Full-bodied, dry, very smoky, pungent – with medicinal notes and some  spicy, malty and fruity notes

#### Bourbons and American Whiskies:

10. “No Rye” whisky with 0 rye gain (pure corn whiskies and "wheaters")
11. “Low Rye” whisky of 10% or less rye grain
12. “Standard Rye” whisky of 10-15% rye grain (classic bourbons)
13. “High Rye” whisky of more 15% rye
14. “Rye” whisky of more than 51% rye (straight rye whiskies)

### 3.3.1 Option A: Recommend Whiskies Similar to User's Chosen Whisky<a class="anchor" id="3_3_1"></a>

```python
def whisky_recommender(whisky_name):
    temp_df_a = df.loc[df['whisky'] == whisky_name]
    predictions_whisky = predict_model(model_gbc, data = temp_df_a)
    predicted_cluster = int(predictions_whisky['Label'])
    cluster_df = df.loc[df['cluster'] == predicted_cluster]
        
    return cluster_df
```

```python
recommended_whiskies_a = whisky_recommender('Amrut Spectrum (Batch 1)')
```

```python
recommended_whiskies_a.sample(5)
```

### 3.3.2 Option B: Recommend Whiskies based on User's Chosen Flavour Profile<a class="anchor" id="3_3_2"></a>

```python
def flavour_recommender(cluster_number):
    temp_df_b = df.loc[df['cluster'] == cluster_number]
    predictions_flavour = predict_model(model_gbc, data = temp_df_b)
        
    return predictions_flavour
```

```python
recommended_whiskies_b = flavour_recommender(9)
```

```python
recommended_whiskies_b.sample(5)
```

## 3.4 Whisky and Food Pairing<a class="anchor" id="3_4"></a>

```python
# load data
df2 = pd.read_csv('data/whisky_food_cleaned.csv')
df2.head()
```

```python
def food_recommender(whisky_cluster):
    if whisky_cluster == 1:
        recommended_food = df2['food'].loc[df2['cluster_1'] == 1]
    elif whisky_cluster == 2:
        recommended_food = df2['food'].loc[df2['cluster_2'] == 1]
    elif whisky_cluster == 3:
        recommended_food = df2['food'].loc[df2['cluster_3'] == 1]
    elif whisky_cluster == 4:
        recommended_food = df2['food'].loc[df2['cluster_4'] == 1]
    elif whisky_cluster == 5:
        recommended_food = df2['food'].loc[df2['cluster_5'] == 1]
    elif whisky_cluster == 6:
        recommended_food = df2['food'].loc[df2['cluster_6'] == 1]
    elif whisky_cluster == 7:
        recommended_food = df2['food'].loc[df2['cluster_7'] == 1]
    elif whisky_cluster == 8:
        recommended_food = df2['food'].loc[df2['cluster_8'] == 1]
    elif whisky_cluster == 9:
        recommended_food = df2['food'].loc[df2['cluster_9'] == 1]
    elif whisky_cluster == 10:
        recommended_food = df2['food'].loc[df2['cluster_10'] == 1]
    elif whisky_cluster == 11:
        recommended_food = df2['food'].loc[df2['cluster_11'] == 1]
    elif whisky_cluster == 12:
        recommended_food = df2['food'].loc[df2['cluster_12'] == 1]
    elif whisky_cluster == 13:
        recommended_food = df2['food'].loc[df2['cluster_13'] == 1]
    else:
        recommended_food = df2['food'].loc[df2['cluster_14'] == 1]
        
    return recommended_food
```

```python
food_recommender(9)
```

## 3.5 Conclusion <a class="anchor" id="3_5"></a>

### Recommendations ###
These basic recommendation systems have given expected results thus far.

### Limitations ###
Since the majority of whiskies are from Scotland, the recommender is more likely to Scotch whiskies compared to whiskies from other regions.

### Future Steps ###
The recommendation systems could be further improved to perhaps display 3 Scotch and 2 non-Scotch whiskies to give the latter a more significant representation when making recommendations to users.

Perhaps a price-related filter could be built in to help users pick whiskies that they find more affordable or that possibly offer more "value-for-money" type of choices.
